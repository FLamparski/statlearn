# Analysing the DFE spending data in the summer of 2014

```python
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
# Import data into data frames. Some normalisation will have to be done. Datasets can be downloaded
# at http://data.gov.uk/dataset/financial-transactions-data-dfe
dfe_aug = pd.read_csv('./DfE_Spend_August_2014.csv', parse_dates=True, dayfirst=True)
dfe_jul = pd.read_csv('./DfE_Spend_July_2014_v2.csv', parse_dates=True, dayfirst=True)
dfe_jun = pd.read_csv('./DfE_Spend_June_2014_-_Final_v1.csv', parse_dates=True, dayfirst=True)
```

```python
# Column names changed in August, need to normalise
dfe_aug = dfe_aug.rename(columns={'Departmental Family': 'Department family', 'Transaction Number': 'Transaction number'})
```

```python
# Get the data for the whole quarter
dfe = pd.concat([dfe_jun, dfe_jul, dfe_aug])
```

```python
dfe.Date = pd.to_datetime(dfe.Date)
```

```python
dfe.groupby('Supplier').size().order(ascending=False)[:20]
```

```python
dfe.groupby('Entity').sum().sort('Amount', ascending=False)
```

```python
dfe[dfe.Entity == 'DEPARTMENT FOR EDUCATION'].groupby('Expense area').sum().sort('Amount', ascending=False)
```

```python
# Let's actually make it time series data
dfe.index = dfe.Date
```

```python
dfe.head()
```

```python
# What are they spending on the most by expense type?
dfe.groupby('Expense type').sum().sort('Amount', ascending=False)
```

```python
# Can we visualise this table?
dfe.groupby('Expense type').sum().sort('Amount', ascending=False).plot(kind='bar', figsize=(16,10), logy=True, ylim=(1e+0,1e+11))
```

```python
# How does amount vs time look like?
dfe.plot(kind='line', x='Date', y='Amount', figsize=(10,8), style='k--')
roll_mean = pd.rolling_mean(dfe.Amount, 20)
roll_mean.plot(kind='line', style='r')
roll_stdev = pd.rolling_std(dfe.Amount, 20)
plt.fill_between(roll_stdev.index, roll_mean-2*roll_stdev, roll_mean+2*roll_stdev, color='b', alpha=0.2)
# ... it doesn't look like anything.
```

```python
# Check if the data is correlated
from pandas.tools.plotting import autocorrelation_plot, lag_plot
```

```python
plt.figure(figsize=(12,10))
autocorrelation_plot(dfe.Amount)
```

```python
plt.figure(figsize=(14,10))
lag_plot(dfe.Amount)
```

```python
# There doesn't seem to be a strong date/amount correlation here.
# So what's the distribution of the spending?
dfe.Amount.plot(kind='kde', logy=True, figsize=(14,10))
```
